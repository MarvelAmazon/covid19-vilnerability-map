# population_dataset

```python
### load the polulation
population_dataset = pd.read_csv('./dataset/hti_pop2019.csv')
population_dataset.isna().sum()
```

```python
display(population_dataset.head())

population_dataset.rename(columns={'adm1_fr':'departement','adm2_fr': 'commune','adm3_fr':'section_communale','adm3code':'adm3_sec_code'},inplace =True)
population_dataset = population_dataset[['adm3_sec_code','departement','commune','section_communale','IHSI_UNFPA_2019_female','IHSI_UNFPA_2019_male','IHSI_UNFPA_2019']]
# convert string to int
# model 
population_dataset['adm3_sec_code'] = population_dataset['adm3_sec_code'].apply(lambda x:re.sub('\D','',x))
population_dataset.adm3_sec_code =population_dataset.adm3_sec_code.astype('int')
display(population_dataset.head())
population_dataset.to_csv('./output/population_dataset.csv')
```

```python
# print of all distict values of seccommunes
print('Total of unique section communales {}'.format(len(set(population_dataset.adm3_sec_code))))
# print of all distict values of seccommunes
print('Total of unique  communes {}'.format(len(set(population_dataset.commune))))
# print of all distict values of departement
print('Total of unique departement {}'.format(len(set(population_dataset.departement))))
```

```python
#### imports the shapefile dataset
sec_communales = gpd.read_file('./hti_boundaries_admi3/hti_boundaries_sections_communales_adm3_cnigs_polygon.shp',encoding ='utf-8')
# renames columns 
sec_communales.rename(columns={'id_section': 'adm3_sec_code','nom_sectio':'section_communale','departemen':'departement','sect_x_lab': 'long','sect_y_lab':'lat'},inplace =True)
# removes non digits characters
sec_communales['adm3_sec_code'] = sec_communales['adm3_sec_code'].apply(lambda x:re.sub('\D','',x))
# convert adm3_sec_code to integer
sec_communales.adm3_sec_code = sec_communales.adm3_sec_code.astype('int')
# selects features needed
sec_communales= sec_communales[['adm3_sec_code','section_communale','commune','departement','long','lat','shape_leng','shape_le_1','shape_le_2','shape_area','geometry']]
sec_communales.head()
```

```python
# print of all distict values of seccommunes
print('Total of unique section communales {}'.format(len(set(sec_communales.adm3_sec_code))))
# print of all distict values of seccommunes
print('Total of unique  communes {}'.format(len(set(sec_communales.commune))))
# print of all distict values of departement
print('Total of unique departement {}'.format(len(set(sec_communales.departement))))
```

```python
# Read the xlsx file with pandas
localtion_health_facilities_hospital = pd.read_excel('./dataset/haiti-healthsites_hdx.xlsx')
# Remove the first column 
localtion_health_facilities_hospital.drop(index = 0, inplace =True)
# display the dataset
localtion_health_facilities_hospital.head()
```

```python
# apply save de dataset (name = haiti-healthsites.csv)
localtion_health_facilities_hospital.to_csv('./output/haiti-healthsites.csv')
```

```python
# print of all distict values of seccommunes
print('Total of unique section communales {}'.format(len(set(localtion_health_facilities_hospital.adm3_fr))))
# print of all distict values of seccommunes
print('Total of unique  communes {}'.format(len(set(localtion_health_facilities_hospital.adm2_fr))))
# print of all distict values of departement
print('Total of unique departement {}'.format(len(set(localtion_health_facilities_hospital.adm1_fr))))

```

```python
## rename columns on the dataset
set(localtion_health_facilities_hospital.adm1_fr)
```

```python
# renames some columns on the dataset
localtion_health_facilities_hospital.rename(columns ={"adm3code": "adm3_sec_code","adm3_en":"section_communale","adm1_fr":"departement","adm2_en":"commune","Catégorie":"categorie","Institution name":"institution_name","Institution_Type":"institution_type","Institution_Code":"institution_code","LatDD":"lat","LongDD":"long"},inplace =True)
# selects columns needed
localtion_health_facilities_hospital =localtion_health_facilities_hospital[['adm3_sec_code','commune','departement','section_communale','institution_name','categorie','institution_type','institution_code','lat','long']]
# remove duplicates

# display data
display(localtion_health_facilities_hospital.head())
# set categorical values
display(set(localtion_health_facilities_hospital.categorie))
display(set(localtion_health_facilities_hospital.institution_type))
```

```python
# improves categorical values to numerics values function
def get_dummies(dataset,columns):
    for column_ in columns:
        data = pd.get_dummies(dataset[column_])
        for column in data.columns:
            dataset[column] = data[column]
    print(dataset.shape)
    return dataset
```

```python
# rename na
localtion_health_facilities_hospital.categorie = localtion_health_facilities_hospital.categorie .fillna('Categories Non spec.')
localtion_health_facilities_hospital.institution_type = localtion_health_facilities_hospital.institution_type.fillna('Type Non spec.')

# apply the fuctions get_dummies
localtion_health_facilities_hospital=get_dummies(localtion_health_facilities_hospital,['categorie','institution_type'])
localtion_health_facilities_hospital
```

```python
# summarizes site by categories and types by commune
summary =localtion_health_facilities_hospital.groupby(['adm3_sec_code']).sum().reset_index().drop_duplicates()
# display total of site by commune
sumary2 = localtion_health_facilities_hospital.groupby(['adm3_sec_code']).size().to_frame().reset_index()
sumary2.rename(columns={0:"Total"},inplace =True)
# merges sumary and sumary 2 to
summary_haiti_healthsites = pd.merge(summary,sumary2,how='inner',on='adm3_sec_code')
# rename non digit characters
summary_haiti_healthsites['adm3_sec_code'] = summary_haiti_healthsites['adm3_sec_code'].apply(lambda x:re.sub('\D','',x))
# convert adm3_sec_code to int
summary_haiti_healthsites.adm3_sec_code =summary_haiti_healthsites.astype('int')
summary_haiti_healthsites
```

```python
# merge population dataset to sec_communales boundaries
sec_communales =pd.merge(sec_communales,population_dataset[['adm3_sec_code','IHSI_UNFPA_2019_female','IHSI_UNFPA_2019_male','IHSI_UNFPA_2019']],how ='left',on='adm3_sec_code')
# merge health sites dataset to sec_communales boundaries
sec_communales = pd.merge(sec_communales,summary_haiti_healthsites,how ='left',on ='adm3_sec_code')
# save to csv file  the merged commune , population and health sites dataset
sec_communales.to_csv('./output/adim3_boundaries_population_and_health_sites.csv')
```
